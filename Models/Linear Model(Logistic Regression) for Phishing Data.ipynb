```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import datasets
```

```python
dataset = pd.read_csv('Datasets/Phishing_dataset.csv')
```

```python
#check head of data set
dataset.head()
```

```python
dataset.info()
```

```python
#info shows that their are 32 columns and 11055 rows of data and no null value is present 
```

```python
#from dataset we dont require the index column for train model
```

```python
dataset = dataset.drop('index',axis=1) #axis = 1 represent column
```

```python
dataset.head()
```

```python
# Now we have to separate labels and data to be tested seperately. 
# In dataset "Result" column describes whether link is legtimate or not.
# theirfore "Result" column is to be tested against labeled data
```

```python
dataset.columns
```

```python
X = dataset.drop('Result',axis=1) 
# X will be our features data so we drop the Result column as its our Target
```

```python
X
```

```python
y = dataset['Result']
# y will be our target
```

```python
y
```

```python
#Split our data into training and testing set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# test_size=0.3 means 30% will be our testing data and 70% training data
# random_state intiallize the shuffling the data, here we set it to 0 for no shuffling
# or you can change it but result may vary         
```

```python
# import Linear regression algorithm to train and test data
from sklearn.linear_model import LogisticRegression
```

```python
#Create a classifier / instantiate model
logmodel = LogisticRegression()
```

```python
# train model by fitting data into it
logmodel.fit(X_train,y_train)
```

```python
# predict the test set
prediction = logmodel.predict(X_test)
```

```python
# predictions are done, now check the stats of predictions against original data

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#confusion matrix
cm = confusion_matrix(y_test, prediction)
print(cm)
```

```python
print(classification_report(y_test,prediction))
```

```python
print(accuracy_score(y_test, prediction))
```

```python
#accuracy 92%
```
